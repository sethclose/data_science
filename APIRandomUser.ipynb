```python
def get_pic (user):
    uri = user.get_picture()
    r=requests.get(uri)
    file_name = './images/' + user.get_full_name().replace(' ','') + '.jpg'
    path=os.path.join(os.getcwd(),file_name)
    with open(path,'wb') as f:           # Open the file (write binary)
        f.write(r.content)     
    #Image.open(path) 
    return file_name
```

```python
pic_file_names = []
for user in users:
    pic_file_names.append(get_pic(user))
pic_file_names
```

```python
def get_users():
    users =[]
     
    for user in RandomUser.generate_users(3):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})

    print(users)
    return pd.DataFrame(users)  
```
