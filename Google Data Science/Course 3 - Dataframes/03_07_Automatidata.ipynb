### Average Tip by Passenger Count

Examine the unique values in the `passenger_count` column.

```python
df.passenger_count.unique()
```

```python
# Calculate mean tips by passenger_count
pass_tips = df.rename(columns={'passenger_count':'Passengers',
                               'tip_amount':'Average Tip'}).groupby(['Passengers']).agg({'Average Tip':'mean'})
pass_tips
```

```python
plt.rcParams["figure.figsize"] = (2,3)
sns.barplot(y=pass_tips['Average Tip'], x=pass_tips.index,
            palette="cividis", hue=pass_tips.index, legend=False)
plt.title('Tips')
plt.xlabel('Number of Passengers')
plt.ylabel(None)
plt.show()
```

```python
# Create bar plot for mean tips by passenger count
mean_tips_by_passenger_count = df.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```

```python
data = mean_tips_by_passenger_count.tail(-1)
pal = sns.color_palette("Greens_d", len(data))
rank = data['tip_amount'].argsort().argsort()
plt.figure(figsize=(5,3))
ax = sns.barplot(x=data.index,
            y=data['tip_amount'],
            palette=np.array(pal[::-1])[rank], hue=data.index)
ax.axhline(df['tip_amount'].mean(), ls='--', color='red', label='global mean')
ax.legend()
plt.title('Mean tip amount by passenger count', fontsize=16);
```

### Total Rides Per Month (Bar)

```python
# Create a month column
try:
    df.insert(4, 'Month', df.tpep_dropoff_datetime.dt.month_name())
except:
    print('Month may already exist.')
```

```python
# Create a day column
try: 
    df.insert(5, 'Day', df.tpep_dropoff_datetime.dt.day_name())
except:
    print('Day may already exist.')    
```

```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
month_counts = df.rename(columns={'ID':'Count'}).groupby('Month').agg({'Count':'size'}).reindex(index=month_order)
plt.rcParams["figure.figsize"] = (4,4)
sns.barplot(y=month_counts.Count, x=month_counts.index,
            palette="viridis", hue=month_counts.index, legend=False)
plt.title('Rides by Month')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

### Plot total ride count by day

Repeat the above process, but now calculate the total rides by day of the week.

```python
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_counts = df.rename(columns={'ID':'Count'}).groupby('Day').agg({'Count':'size'}).reindex(index=day_order)
plt.rcParams["figure.figsize"] = (4,4)
sns.barplot(y=day_counts.Count, x=day_counts.index,
            palette="viridis", hue=day_counts.index, legend=False)
plt.title('Rides by Day')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

**Plot total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
df.head(1)
```

```python
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_counts = df.rename(columns={'total_amount':'Revenue'}).groupby('Day').agg({'Revenue':'sum'}).reindex(index=day_order)
plt.rcParams["figure.figsize"] = (4,4)
sns.barplot(y=day_counts.Revenue, x=day_counts.index,
            palette="BrBG", hue=day_counts.index, legend=False)
plt.title('Revenue by Day')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

**Plot total revenue by month**

```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
month_counts = df.rename(columns={'total_amount':'Revenue'}).groupby('Month').agg({'Revenue':'sum'}).reindex(index=month_order)
plt.rcParams["figure.figsize"] = (4,3)
sns.barplot(y=month_counts.Revenue, x=month_counts.index,
            palette="twilight_r", hue=month_counts.index, legend=False)
plt.title('Rides by Month')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

### Plot mean trip distance by drop-off location

```python
locs = df.rename(columns={'trip_distance':'Distance'}).groupby('DOLocationID').agg({'Distance':'mean'}).sort_values(by='Distance', ascending=False).reset_index()
plt.rcParams["figure.figsize"] = (12,3)
fig = sns.barplot(x=locs.index, y=locs.Distance, palette="twilight_r", hue=locs.index, legend=False)
plt.title('Average Distance by Location')
fig.set(xticklabels=[])
fig.set(xticks=[])
plt.xlabel('')
plt.ylabel('Miles')
plt.show()
```
